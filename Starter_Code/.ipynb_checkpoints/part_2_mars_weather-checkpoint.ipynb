## Deliverable 2: Scrape and Analyze Mars Weather Data

```python
# Import relevant libraries
from splinter import Browser
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import pandas as pd
```

```python
browser = Browser('chrome')
```

### Step 1: Visit the Website

Use automated browsing to visit the [Mars Temperature Data Site](https://static.bc-edx.com/data/web/mars_facts/temperature.html). Inspect the page to identify which elements to scrape.

   > **Hint** To identify which elements to scrape, you might want to inspect the page by using Chrome DevTools to discover whether the table contains usable classes.


```python
# Visit the website
# https://static.bc-edx.com/data/web/mars_facts/temperature.html
url = "https://static.bc-edx.com/data/web/mars_facts/temperature.html"
browser.visit(url)
```

### Step 2: Scrape the Table

Create a Beautiful Soup object and use it to scrape the data in the HTML table.

Note that this can also be achieved by using the Pandas `read_html` function. However, use Beautiful Soup here to continue sharpening your web scraping skills.

```python
# Create a Beautiful Soup Object
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Extract all rows of data
table = soup.find('table', class_='table')  # Assuming the table has class 'table'
rows = table.find_all('tr')  # Find all rows in the table
```

### Step 3: Store the Data

Assemble the scraped data into a Pandas DataFrame. The columns should have the same headings as the table on the website. Here’s an explanation of the column headings:

* `id`: the identification number of a single transmission from the Curiosity rover
* `terrestrial_date`: the date on Earth
* `sol`: the number of elapsed sols (Martian days) since Curiosity landed on Mars
* `ls`: the solar longitude
* `month`: the Martian month
* `min_temp`: the minimum temperature, in Celsius, of a single Martian day (sol)
* `pressure`: The atmospheric pressure at Curiosity's location

```python
# Create an empty list
data_rows = []

# Loop through the scraped data to create a list of rows
for row in rows[1:]:  # Skipping the first row (header)
    # Extract the data for each cell in the row
    cells = row.find_all('td')
    row_data = [cell.text.strip() for cell in cells]  # Extract text from each cell
    data_rows.append(row_data)  # Append to the list of rows
```

```python
# Create a Pandas DataFrame by using the list of rows and a list of the column names
columns = ['id', 'terrestrial_date', 'sol', 'ls', 'month', 'min_temp', 'pressure']
mars_weather_df = pd.DataFrame(data_rows, columns=columns)
```

```python
# Confirm DataFrame was created successfully
print(mars_weather_df.head())
```

### Step 4: Prepare Data for Analysis

Examine the data types that are currently associated with each column. If necessary, cast (or convert) the data to the appropriate `datetime`, `int`, or `float` data types.

  > **Hint** You can use the Pandas `astype` and `to_datetime` methods to accomplish this task.


```python
# Examine data type of each column
print(mars_weather_df.dtypes)
```

```python
# Change data types for data analysis
mars_weather_df['terrestrial_date'] = pd.to_datetime(mars_weather_df['terrestrial_date'])
mars_weather_df['sol'] = pd.to_numeric(mars_weather_df['sol'])
mars_weather_df['ls'] = pd.to_numeric(mars_weather_df['ls'])
mars_weather_df['month'] = pd.to_numeric(mars_weather_df['month'])
mars_weather_df['min_temp'] = pd.to_numeric(mars_weather_df['min_temp'])
mars_weather_df['pressure'] = pd.to_numeric(mars_weather_df['pressure'])
```

```python
# Confirm type changes were successful by examining data types again
print(mars_weather_df.dtypes)
```

### Step 5: Analyze the Data

Analyze your dataset by using Pandas functions to answer the following questions:

1. How many months exist on Mars?
2. How many Martian (and not Earth) days worth of data exist in the scraped dataset?
3. What are the coldest and the warmest months on Mars (at the location of Curiosity)? To answer this question:
    * Find the average the minimum daily temperature for all of the months.
    * Plot the results as a bar chart.
4. Which months have the lowest and the highest atmospheric pressure on Mars? To answer this question:
    * Find the average the daily atmospheric pressure of all the months.
    * Plot the results as a bar chart.
5. About how many terrestrial (Earth) days exist in a Martian year? To answer this question:
    * Consider how many days elapse on Earth in the time that Mars circles the Sun once.
    * Visually estimate the result by plotting the daily minimum temperature.


```python
# 1. How many months are there on Mars?
unique_months = mars_weather_df['month'].nunique()
print(f'There are {unique_months} months on Mars.')
```

```python
# 2. How many sols (Martian days) worth of data are there?
total_sols = mars_weather_df['sol'].nunique()
print(f'There are {total_sols} Martian days worth of data.')
```

```python
# 3. What is the average minimum temperature by month?
avg_min_temp_by_month = mars_weather_df.groupby('month')['min_temp'].mean()
print("Average minimum temperature by month:\n", avg_min_temp_by_month)
```

```python
# Plot the average minimum temperature by month
avg_min_temp_by_month.plot(kind='bar', title='Average Minimum Temperature by Month on Mars', ylabel='Min Temp (°C)')
plt.show()
```

```python
# Identify the coldest and hottest months in Curiosity's location by sorting the previous graph
coldest_month = avg_min_temp_by_month.idxmin()
hottest_month = avg_min_temp_by_month.idxmax()
print(f'The coldest month is month {coldest_month}.')
print(f'The hottest month is month {hottest_month}.')
```

```python
# 4. What is the average pressure by month?
avg_pressure_by_month = mars_weather_df.groupby('month')['pressure'].mean()
print("Average pressure by month:\n", avg_pressure_by_month)
```

```python
# Plot the average pressure by month
avg_pressure_by_month.plot(kind='bar', title='Average Atmospheric Pressure by Month on Mars', ylabel='Pressure (Pa)')
plt.show()
```

```python
# Identify the lowest and highest pressure months in Curiosity's location by sorting the previous graph
lowest_pressure_month = avg_pressure_by_month.idxmin()
highest_pressure_month = avg_pressure_by_month.idxmax()
print(f'The month with the lowest pressure is month {lowest_pressure_month}.')
print(f'The month with the highest pressure is month {highest_pressure_month}.')
```

```python
# 5. How many terrestrial (Earth) days are there in a Martian year?
# Visually estimate the result by plotting the daily minimum temperature of each observation in the data set.
mars_weather_df = mars_weather_df.sort_values(by='terrestrial_date')
plt.figure(figsize=(10,6))
plt.plot(mars_weather_df['terrestrial_date'], mars_weather_df['min_temp'], label='Min Temp')
plt.xlabel('Terrestrial Date')
plt.ylabel('Minimum Temperature (°C)')
plt.title('Daily Minimum Temperature on Mars over Time')
plt.grid(True)
plt.show()
```

#### Minimum Temperature

The average minimum temperature varies across the 12 Martian months, with a distinct pattern:

Coldest Month: Month 3 has the lowest average minimum temperature at approximately -83.3°C. This could correspond to the Martian winter in the hemisphere where Curiosity is located.
Hottest Month: Month 8 is the warmest, with an average minimum temperature of about -68.4°C. Even though this is relatively warm for Mars, it still represents freezing conditions by Earth standards.
Temperature Range: The temperature fluctuates between -83.3°C and -68.4°C, showing a significant variation in Martian surface temperature, likely due to seasonal changes.

#### Atmospheric Pressure

The average atmospheric pressure also shows significant variation across Martian months:

Highest Pressure: Month 9 has the highest average pressure at about 913 Pa. This may correspond to the Martian winter in the southern hemisphere, where CO₂ from the polar ice cap sublimates into the atmosphere, increasing atmospheric pressure.
Lowest Pressure: Month 6 has the lowest average pressure at around 745 Pa. This could represent a phase when the polar ice caps begin to re-freeze, removing CO₂ from the atmosphere and lowering the pressure.
Pressure Range: The pressure fluctuates between 745 Pa and 913 Pa, which is extremely low compared to Earth's average sea-level pressure of about 101,325 Pa, but normal for Mars.

#### Year Length

Based on the Daily Minimum Temperature plot over terrestrial dates, we can observe clear cyclical temperature variations, indicating Martian seasonal changes. From this plot:

It is visually evident that one complete cycle, representing one Martian year, spans roughly 687 Earth days, consistent with scientific knowledge about Mars' orbital period around the Sun.
The temperature curve repeats itself approximately every 1.9 Earth years, which aligns well with the actual Martian year length.
